```python
right_img = cv2.imread("右.png")
left_img = cv2.imread("左.png")
right_img = cv2.cvtColor(right_img, cv2.COLOR_BGR2RGB)
left_img = cv2.cvtColor(left_img, cv2.COLOR_BGR2RGB)
```

```python
#画像サイズが違うので同じサイズに変換
#グレースケール化しても良い
right_img = cv2.resize(right_img, dsize=(1140, 1128))
#right_img = cv2.cvtColor(right_img, cv2.COLOR_BGR2GRAY)
left_img = cv2.resize(left_img, dsize=(1140, 1128))
#left_img = cv2.cvtColor(left_img, cv2.COLOR_BGR2GRAY)
#plt.gray()
```

```python
#差分を求める
img = cv2.absdiff(right_img,left_img)
```

```python
fig = plt.figure(figsize=(15,15))
X = 1
Y = 3
ax1 = fig.add_subplot(X, Y, 1)
ax1.set_title("img1",fontsize=20)
plt.imshow(right_img)

ax2 = fig.add_subplot(X, Y, 2)
#タイトルの設定
ax2.set_title("img2",fontsize=20)
plt.imshow(left_img)

ax3 = fig.add_subplot(X, Y, 3)
#タイトルの設定
ax3.set_title("img3",fontsize=20)
plt.imshow(img)
```
